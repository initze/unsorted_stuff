```python
for idx in ['NDVI', 'NDMI', 'NDWI', 'TCB',  'TCG', 'TCW', 'NBR']:
    try:
        cols = df.columns[df.columns.str.contains(idx)]
        df[cols].to_csv(r'./Nyupa_{i}.csv'.format(i=idx))
    except:
        continue
```
